### Retrouver le centre $C$ du berceau

Il est nécesssaire de trouver le centre $C$ du cercle passant au mieux des points pour obtenir l'angle d'oscillation du centre $G$ du rotor $\mathbf{2}$ par rapport au berceau $\mathbf{0}$.

Le code ci-dessous permet de le faire par la <a href="https://fr.wikipedia.org/wiki/R%C3%A9gression_circulaire" target="_blank">régression cette fois-ci circulaire par la méthode des moindres carrés</a>.
    
<p class="question"><br/>
    Il vous suffit de compiler le code ci-dessous. Il vous retourne<br/>
    <span style="padding:0 0 0 25px;">&bull; l'abcisse du point $C$ sous le nom de variable <span class="code">xC</span></span><br/>
    <span style="padding:0 0 0 25px;">&bull; l'ordonnée du point $C$ sous le nom de variable <span class="code">yC</span></span><br/>
    <span style="padding:0 0 0 25px;">&bull; le diamètre du cercle sous le nom de variable <span class="code">D</span></span><br/>
</p>

```python
# Compiler le code

from numpy import sqrt

def cercle(x,y):
    Sxi, Syi, Sxiyi, Sxi2, Syi2, Sxiyi2, Sxi2yi, Sxi3, Syi3 = 0, 0, 0, 0, 0, 0, 0, 0, 0

    n = len(x)

    for i in range(n):
        Sxi    += x[i]
        Syi    += y[i]
        Sxiyi  += x[i]*y[i]
        Sxi2   += x[i]**2
        Syi2   += y[i]**2
        Sxiyi2 += x[i]*(y[i]**2)
        Sxi2yi += (x[i]**2)*y[i]
        Sxi3   += x[i]**3
        Syi3   += y[i]**3

    a = Sxi2-((Sxi**2)/n)
    b = Sxiyi-(Sxi*Syi/n)
    c = Syi2-((Syi**2)/n)
    d = (Sxi3+Sxiyi2-(Sxi*(Sxi2+Syi2)/n))/2
    e = (Syi3+Sxi2yi-(Syi*(Sxi2+Syi2)/n))/2

    xC = (c*d-b*e)/(a*c-b**2)
    yC = (a*e-b*d)/(a*c-b**2)
    R = sqrt(((Sxi2+Syi2-2*(xC*Sxi+yC*Syi))/n)+xC**2+yC**2)

    return xC, yC, 2*R

xC, yC, D = cercle(xG,yG)

txt  = '<p class="compilation"><br/>'
txt += f'Le cercle a pour centre C d’abscisse <span class="gris">xC</span> de <span class="gris">{1000*xC:.3f} mm</span> et d’ordonnée <span class="gris">yC</span> de <span class="gris">{1000*yC:.3f} mm</span> et de diamètre D de {D:.3f} m.<br/>'
txt += '<br/>'

def cicularite(xC, yC, R, xG, yG):
    listeEcarts = []

    for i in range(len(xG)):
        Ei = sqrt((xG[i]-xC)**2+(yG[i]-yC)**2) - R
        listeEcarts.append(Ei)

    Emax, Emin = max(listeEcarts), min(listeEcarts)

    return Emin, Emax

Emin, Emax = cicularite(xC, yC, D/2, xG, yG)

txt += f'La circularité minimum est de {1000*Emin:.3f} mm sur un rayon de {1000*D/2:.3f} mm.<br/>'
txt += f'La circularité maximum est de {1000*Emax:.3f} mm sur un rayon de {1000*D/2:.3f} mm.<br/>'
txt += '<br/>'
txt += f'Conclusion : La circularité est très bonne mais le rayon théorique est de 1400 mm au lieu de {1000*D/2:.3f} mm. Gardons cela en tête !<br/>'
txt += '</p>'
html(txt)
```

```python

def thetaEDCr(Cr) :
    Ne=10**6
    
    Te=t[-1]/Ne
    thetaED=[theta[0]]
    dthetaED=[0]
    ddthetaED=[(-np.sign(dthetaED[0])*Cr*r-m*g*r**2*np.sin(thetaED[0]))/((J+m*r**2)*(R-r))]
    for i in range(Ne) :
        thetaED.append(thetaED[i]+Te*dthetaED[i])
        dthetaED.append(dthetaED[i]+Te*ddthetaED[i])
        ddthetaED.append( (-np.sign(dthetaED[i+1])*Cr*r-m*g*np.sin(thetaED[i+1])*r**2)/((J+m*r**2)*(R-r)) )
    return ([Te*i for i in range(Ne+1)],thetaED)
tED,thetaED=thetaEDCr(0)
```

## Détermination de la résistance au roulement $C_r$ (activité intéréssante cependant hors de la problématique)

Nous vous proposons une fonction <span class="code">ecartCumul(t, theta, tED, thetaED)</span> qui calcule l'écart entre les termes des signaux <span class="code">theta</span> avec ces instants <span class="code">t</span> et <span class="code">thetaED</span> avec ces instants <span class="code">tED</span>.

On veillera toutefois à ce que la longueur des éléments de <span class="code">theta</span> soit plus petite que celle de <span class="code">thetaED</span> (rapport de 10 minimum).

```python
# Compiler le code

def ecartCumul(t, theta, tED, thetaED):
    assert len(theta) <= len(thetaED) / 10

    # Eliminer les points inutiles
    thetaEDNew = [None] * len(t)
    indTED = 0
    for indT in range(len(t)):
        while indTED < len(tED) and tED[indTED] <= t[indT]:
            thetaEDNew[indT] = thetaED[indTED]
            indTED += 1

    # Comparer les courbes
    res = 0
    for i in range(len(t)):
        if thetaEDNew[i] == None: # Précaution
            res += 0
        elif theta[i] >= 0:
            res += thetaEDNew[i] - theta[i]
        else:
            res -= thetaEDNew[i] - theta[i]

    return res

txt  = '<p class="compilation"><br/>'
txt += f'Pour l’exemple ici de $C_r = 0$ $ N \\cdot m$, l’écart de ${ecartCumul(t, theta, tED, thetaED):.1f}$ $radians$.'
txt += '</p>'
html(txt)
```

<p class="question"><br/>
    En utilisant la dichtomie, évaluer la résistance au roulement $Cr$ avec une erreur inférieur à $10^{-5} N \cdot m$.
</p>

```python
def dicho(t_ref,theta_ref,Cr_min=0,Cr_max=0.03,ED=thetaEDCr) :
    while Cr_max-Cr_min>10**-7 :
        Cr=(Cr_max+Cr_min)/2
        tED,thetaED=ED(Cr)
        dist=ecartCumul(t_ref,theta_ref,tED,thetaED)
        if dist==0 :
            return Cr
        elif dist<0 :
            Cr_max=Cr
        else :
            Cr_min=Cr
    return Cr
Cr=dicho(t[:1000],theta[:1000])
print(Cr)
```
